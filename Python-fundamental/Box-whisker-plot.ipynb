# Box-Whisker plot

- A box plot, also known as a box-and-whisker plot, is a graphical representation of a dataset that displays the distribution of numerical data through quartiles. It provides a visual summary of the minimum, first quartile, median, third quartile, and maximum values of a dataset.
- In a box plot, a rectangular box is drawn, which represents the interquartile range (IQR) of the data. The box is divided into two parts by a vertical line at the median. The lower part of the box represents the first quartile (25th percentile), while the upper part represents the third quartile (75th percentile). The whiskers, represented by lines extending from the box, typically show the minimum and maximum values within a certain range, excluding outliers.
- Outliers, which are data points that significantly deviate from the rest of the dataset, are often represented as individual points beyond the whiskers or as separate symbols. They can provide insights into potential anomalies or extreme values in the data.
- Box plots are useful for comparing distributions, identifying skewness, detecting outliers, and gaining a general understanding of the spread and central tendency of a dataset. They are commonly used in statistical analysis and data visualization to summarize and interpret numerical data.

> - A box plot is constructed using the minimum and maximum quartiles (Q1 and Q3) and the median of a data set.
> 
> 
> 
> - The horizontal axis covers all possible data values.
> - The box part of a box-and-whisker plot covers the middle 50% of the values in the data set. This 50% of the data lies between Q1 and Q3 and its range is the interquartile range: $IQR = Q3-Q1$
> 
> - The whiskers each cover 25% of the data values.
>   - The lower whisker covers all the data values from the minimum value up to Q1, that is, the lowest 25% of data values.
>   - The upper whisker covers all the data values between Q3 and the maximum value, that is, the highest 25% of data values.
> - The median sits within the box and represents the center of the data. 50% of the data values lie above the median and 50% lie below the median.
> Outliers, or extreme values in a data set, are usually indicated on a box-and-whisker plot by the “star” symbol. If there is one or more outliers in a data set, for the purpose of drawing a box-and-whisker plot, we take the minimum and maximum to be the minimum and maximum values of the data set excluding the outliers.

1. **Basic Box Plot:** This example shows how to create a basic box plot with random data:

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Create random data
data = [np.random.normal(0, 1, 100) for _ in range(3)]

# Create a box plot
plt.boxplot(data)

# Add labels and title
plt.xlabel('Groups')
plt.ylabel('Values')
plt.title('Basic Box Plot')

# Show the plot
plt.show()
```

2. **Horizontal Box Plot:** You can also create a horizontal box plot by specifying `vert=False`:

```python
import matplotlib.pyplot as plt
import numpy as np

# Create random data
data = [np.random.normal(0, 1, 100) for _ in range(3)]

# Create a horizontal box plot
plt.boxplot(data, vert=False)

# Add labels and title
plt.xlabel('Values')
plt.ylabel('Groups')
plt.title('Horizontal Box Plot')

# Show the plot
plt.show()
```

3. **Grouped Box Plot:** You can create grouped box plots to compare multiple datasets side by side:

```python
import matplotlib.pyplot as plt
import numpy as np

# Create three sets of random data
data1 = np.random.normal(0, 1, 100)
data2 = np.random.normal(2, 1, 100)
data3 = np.random.normal(4, 1, 100)

# Combine the data into a list of lists
data = [data1, data2, data3]

# Create a grouped box plot
plt.boxplot(data, labels=['Group 1', 'Group 2', 'Group 3'])

# Add labels and title
plt.xlabel('Groups')
plt.ylabel('Values')
plt.title('Grouped Box Plot')

# Show the plot
plt.show()
```

4. **Customizing Box Plots:** You can customize the appearance of box plots by specifying various parameters such as colors, styles, and notches:

```python
import matplotlib.pyplot as plt
import numpy as np

# Create random data
data = [np.random.normal(0, 1, 100) for _ in range(3)]

# Create a box plot with customizations
plt.boxplot(data, notch=True, patch_artist=True, boxprops=dict(facecolor='lightblue'))

# Add labels and title
plt.xlabel('Groups')
plt.ylabel('Values')
plt.title('Customized Box Plot')

# Show the plot
plt.show()
```
